## 데이터 시각화
* 데이터를 관찰하는 과정에서 선택사항이 아니라 반드시 거쳐야 하는 필수 과정
* 숫자와 텍스트만으로 구성된 통계결과는 지루/따분
* 그래프와 그림형태로 제시하면 데이터의 관계구분 용이, 열정적인 반응 동반

## R 기본 그래프 함수
* plot, hist, boxplot, barplot, pie, stem

```r
# 표준정규분포를 따르는 난수 15개 생성하고
# 그에 따른 그래프를 작성해서 시각화
# 정규분포 : 수집한 자료의 형태가 평균 0, 표준편차가 1인 형태
# rnorm(갯수,  mean=0, sd=1) : -1 ~ 1 사이의 난수 생성 함수
x <- rnorm(15, mean=0, sd=1)
```

```r
x
```

```r
summary(x)
```

```r
# 산점도 : plot(x, y, 옵션)
plot(x)     # 변수가 1개일 때는 자동으로 y축에 할당
```

```r
# 막대그래프 : barplot(x, y, 옵션)
barplot(x)
```

```r
# 히스토그램 : hist(s, 구간, 옵션)
hist(x)
```

```r
# 상자수염그래프 ㅣ boxplot(x, 옵션)
boxplot(x)
```

```r
# 줄기잎그래프 : stem(x, 옵션)
stem(x)
```

## 그래프 다양한 설정 보기
+ main : 제목
+ sub : 부제목
+ xlab/ylab : 축 제목
+ xlim/ylim : 축값 범위
+ col : 그래프 색상
+ type : 그래프 유형

```r
plot(heights, main='학생 10명 키', sub='class A',
     xlab='numbers', ylim=c(140,190), col='red', type='b', pch=3)
grid()     # 그래프 배경에 모눈 표시
# 산점도 유형(type) : p(점), l(선), b(점,선)
# 산점도 점문자 유형(pch)
# 산점도 선 유형(lwd)
# ?pch
```

## 그래프 눈금 세부 조정하기
+ 그래프 그릴 때 axes 속성을 FALSE로 지정
+ axis 함수로 세부 조정
+ axis(축지정, 눈금지정)

```r
plot(heights, main='student heights', sub='class A',
     xlab='numbers', col='blue', pch=3, axes=F)
axis(side=1, at=1:10)                    # x축 설정
axis(side=2, at=seq(140,185,5), las=2)   # y축 설정
box()     # 그래프 테두리 표시
grid()
```

## 그래프 텍스트 추가하기
+ 특정 x, y위치에 글자 표시
+ text(y좌표, 글자크기, 글자위치, 글자내용)
+ adj : 음수 - 오른쪽 위, 양수 - 왼쪽 아래

```r
plot(heights, col='red')
text(heights, cex=0.7, adj=c(0, -0.2), lab=heights)
grid()
```

```r
# 두 약물에 따른 환자반응 데이터
dose <- c(20,30,40,45,60)     # 나이
drugA <- c(16,20,27,40,60)    # 효과정도
drugB <- c(15,18,25,31,40)
```

```r
# A 약물에 따른 환자반응 - 눈금조정, 텍스트 추가
plot(dose, drugA, main='drugA', sub='',
     xlab='age', ylab='effect', col='red',  axes=F)
axis(side=1, at=seq(20,60,5))
axis(side=2, at=seq(15,60,5))
text(dose, drugA, cex=1, adj=c(0, -0.2), lab=drugA)
box()  
grid()
```

```r
# 범주형 데이터를 토대로 붓꽃의 품종을 색상으로 표현
# 1:black, 2:red, 3:green, 4:blue
plot(iris$Petal.Length, iris$Petal.Width, 
     col=as.integer(iris$Species)+1)
```

```r
# 범주형 데이터를 토대로 붓꽃의 품종을 색상과 기호로 나타낼 수 있음
plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), col=as.integer(iris$Species)+1)
```

## 범례 표시하기
+ legend(위치, 값, 기호, 색상)
+ 위치: top, bottom, right, left, center

```r
plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), col=as.integer(iris$Species)+1)
legend('topleft', 
       c('setosa','versicolor','verginica'), 
       pch=c('○','△','+'),
       col=c('red','green','blue'))
```

```r
# 자동으로 텍스트, 기호, 색상 지정
plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), col=as.integer(iris$Species)+1)
legend('bottomright', 
       levels(iris$Species), 
       pch=unique(as.integer(iris$Species)),
       col=unique(as.integer(iris$Species)+1))
```

## 2개의 산점도를 동시에 출력
+ plot함수와 lines 함수를 번갈아가며 사용
+ 이런 경우 반드시 범례를 출력할 것!

```r
# A,B 약물에 따른 환자반응
plot(dose, drugA, col='red', type='b', pch=1, ylab='drugs')
lines(dose, drugB, col='blue', type='b', pch=2)
legend('topleft', c('drugA','drugB'), pch=c('1','2'), col=c(2,4))
```

## 벡터들로 데이터프레임 생성하기
+ data.frame(벡터1, 벡터2, 벡터3, ... )

```r
# 매니저별 성별/국적별 나이에 대한 산점도 작성
manager <- c(1,2,3,4,5)
country <- c('US','US','US','UK','UK')
gender <- c('M','F','F','M','F')
age <- c(32,45,25,39,99)
```

```r
managers <- data.frame(manager, country, gender, age)
str(managers)
```

```r
# 국적별 나이
plot(managers$age, 
     pch=as.integer(managers$country),
     col=as.integer(managers$country)+1)
legend('topleft', levels(managers$country),
      pch=unique(as.integer(managers$country)),
      col=c('green','red'))
```

```r
# 성별 나이
plot(managers$age, 
     pch=as.integer(managers$gender),
     col=as.integer(managers$gender)+1)
legend('topleft', levels(manager$gender),
      pch=unique(as.integer(managers$gender)),
      col=c('green','red'))
```

```r
# 성별 나이
plot(managers$age, 
     pch=c(1,2),
     col=as.integer(managers$gender)+1)
legend('topleft', levels(manager$gender),
      pch=unique(c(2,1)),
      col=c('red','green'))
```
